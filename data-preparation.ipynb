```python
import pandas as pd
import os
import shutil
```

```python
df = pd.read_csv('./data/ISIC_2019_Training_GroundTruth.csv')
```

```python
df.head()
```

```python
mel_list=df[df['MEL'] == 1.0]['image'].tolist()
len(mel_list)
```

```python
source = './data/ISIC_2019_Training_Input/'
dest = './data/melanoma/train/'

files = os.listdir(source)

for file in files:
    name = file.split('.')[0]
    if name in mel_list:
        file_name = "melanoma_" + str(file)
        shutil.copy(source + file, dest + file_name)
```
